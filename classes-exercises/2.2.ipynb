```python
#importando o PulP
import pulp

#Definindo o modelo
model = pulp.LpProblem('2.2', sense = pulp.LpMaximize)

#Adicionando as variáveis
x = pulp.LpVariable.dicts(indexs=[1,2], cat=pulp.LpContinuous, lowBound=0, name='x')

#Adicionando as restrições
model.addConstraint(x[1] <= 40, name='restricao_1')
model.addConstraint(x[2] <= 30, name='restricao_2')
model.addConstraint(2*x[1] + 3*x[2] <= 120, name='restricao_3')

#Funcao Objetivo
model.setObjective(100 * x[1] + 150 * x[2])

#Optimizar
model.solve()

#Obter e imprimir a solução
x_sol = {i: x[i].value() for i in [1, 2 ]}
print(f'x = {x_sol}')
```
